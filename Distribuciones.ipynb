```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Cargamos data de seaborn
tips = sns.load_dataset("tips")
tips.head() #visualizamos la data
```

```python
#creamos gráfico con seaborn, data, respecto a tip y con 15 barras.
sns.histplot(data=tips, x='tip', bins=15)
plt.show()
```

```python
#creamos gráfico con seaborn, data, respecto a tip y con 15 barras.
#Agregamos comportamiento acumulativo, suma de valores anteriores
sns.histplot(data=tips, x='tip', bins=15, cumulative=True)
plt.show()
```

```python
#creamos gráfico con seaborn, variables data-tip-sex y con 15 barras.
sns.histplot(data=tips, x='tip', bins=15, cumulative=False, hue='sex')
plt.show()
```

```python
#creamos gráfico con seaborn, variables data-tip-sex y con 15 barras.
#Con estadísticas stat=frequency
sns.histplot(data=tips, x='tip', bins=15, cumulative=False, hue='sex', stat='frequency')
plt.show()
```

```python
#creamos gráfico con seaborn, variables data-tip-sex y con 15 barras.
#Con estadísticas stat=probability 0-1
sns.histplot(data=tips, x='tip', bins=15, cumulative=False, hue='sex', stat='probability')
plt.show()
```

```python
#creamos gráfico con seaborn, variables data-tip-sex y con 15 barras.
#Con estadísticas stat=percent
sns.histplot(data=tips, x='tip', bins=15, cumulative=False, hue='sex', stat='percent')
plt.show()
```

```python
#creamos gráfico con seaborn, variables data-tip-sex y con 15 barras.
#Con estadísticas stat=density
sns.histplot(data=tips, x='tip', bins=15, cumulative=False, hue='sex', stat='density')
plt.show()
```

```python
#creamos gráfico con seaborn, variables data-tip-sex y con 15 barras.
#Con estadísticas stat=count
sns.histplot(data=tips, x='tip', bins=15, cumulative=False, hue='sex', stat='count')
plt.show()
```

```python
#creamos gráfico con seaborn, variables data-tip-sex y con 15 barras.
#Con estadísticas stat, multiple=stack uno encima del otro
sns.histplot(data=tips, x='tip', bins=15, cumulative=False, hue='sex', stat='count', multiple='stack')
plt.show()
```

```python
#creamos gráfico con seaborn, variables data-tip-sex y con 15 barras.
#Con estadísticas stat, multiple=layer
sns.histplot(data=tips, x='tip', bins=15, cumulative=False, hue='sex', stat='count', multiple='layer')
plt.show()
```

```python
#creamos gráfico con seaborn, variables data-tip-sex y con 15 barras.
#Con estadísticas stat, multiple=dodge uno al lado del otro
sns.histplot(data=tips, x='tip', bins=15, cumulative=False, hue='sex', stat='count', multiple='dodge')
plt.show()
```

```python
#creamos gráfico con seaborn, variables data-tip-sex y con 15 barras.
#Con estadísticas stat, multiple=fill
sns.histplot(data=tips, x='tip', bins=15, cumulative=False, hue='sex', stat='count', multiple='fill')
plt.show()
```

```python
#creamos gráfico de densidad con seaborn, variables data-tip-sex
sns.kdeplot(data=tips, x='tip', hue='sex')
plt.show() #visualizamos el gráfico
```

```python
#creamos gráfico de densidad con seaborn, variables data-tip-sex
#acumulativo,
sns.kdeplot(data=tips, x='tip', hue='sex', cumulative=True)
plt.show() #visualizamos el gráfico
```

```python
#creamos gráfico de densidad con seaborn, variables data-tip-sex
#fill,
sns.kdeplot(data=tips, x='tip', hue='sex', cumulative=False, fill=True)
plt.show() #visualizamos el gráfico
```

```python
#creamos gráfico de densidad con seaborn, variables data-tip-sex
#fill, bw_adjust ajusta gráfica  parámetros
sns.kdeplot(data=tips, x='tip', hue='sex', cumulative=False, fill=True, bw_adjust=1)
plt.show() #visualizamos el gráfico
```

```python
#creamos gráfico escalonado con seaborn, variables data-tip-sex
sns.ecdfplot(data=tips, x='tip', hue='sex')
plt.show() #visualizamos el gráfico
```

```python
#creamos gráfico escalonado con seaborn, variables data-tip-sex
sns.ecdfplot(data=tips, x='tip', hue='sex', stat='count')
plt.show() #visualizamos el gráfico
```

```python
#creamos gráfico distribución con seaborn, variables data-tip-sex
sns.displot(data=tips, x='tip', hue='sex')
plt.show() #visualizamos el gráfico
```

```python
#creamos gráfico distribución de densidad kde, con seaborn, variables data-tip-sex
sns.displot(data=tips, x='tip', hue='sex', kind='kde')
plt.show() #visualizamos el gráfico
```

```python
#creamos gráfico distribución hist stack con seaborn, variables data-tip-sex
sns.displot(data=tips, x='tip', hue='sex', kind='hist', multiple='stack')
plt.show() #visualizamos el gráfico
```
