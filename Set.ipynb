```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Graficamos en seaborn gráfico barra variable x y
sns.barplot(x=['A','B','C'], y=[1,3,2])
plt.show()
```

```python
#Graficamos en seaborn gráfico barra variable x y
#Implementamos propiedades set, para emplear estilos y modificar gráfico
sns.set(style='dark')
sns.barplot(x=['A','B','C'], y=[1,3,2])
plt.show()
```

```python
#Graficamos en seaborn gráfico barra variable x y
#Implementamos propiedades set, emplear estilos, y modificar gráfico, paleta colores no estilo
sns.set(style='dark', palette='dark')
sns.barplot(x=['A','B','C'], y=[1,3,2])
plt.show()
```

```python
#Graficamos en seaborn gráfico barra variable x y
#Implementamos propiedades set, emplear estilos, y modificar gráfico, paleta colores no estilo,
#Cambiamos Fuente
sns.set(style='dark', palette='dark', font='fantasy', font_scale=1)
sns.barplot(x=['A','B','C'], y=[1,3,2])
plt.show()
```
