```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")
tips.head(2)
```

```python
#Unimos dos distintar gráficas en uno solo
sns.jointplot(data=tips, x='total_bill', y='tip')
plt.show()
```

```python
#Unimos dos distintar gráficas en uno solo
#adicionamos variable sex
sns.jointplot(data=tips, x='total_bill', y='tip', hue='sex')
plt.show()
```

```python
#Unimos dos distintar gráficas en uno solo
#adicionamos variable sex, tipo histograma
sns.jointplot(data=tips, x='total_bill', y='tip', hue='sex', kind='hist')
plt.show()
```

```python
#Unimos dos distintar gráficas en uno solo
#adicionamos variable sex, tipo kde
sns.jointplot(data=tips, x='total_bill', y='tip', hue='sex', kind='kde')
plt.show()
```

```python
#Unimos dos distintar gráficas en uno solo
#adicionamos variable sex, tipo histograma, agregamos valores a histogramas ejes x y y
sns.jointplot(data=tips, x='total_bill', y='tip', hue='sex', kind='hist', marginal_ticks=True)
plt.show()
```

```python
#Unimos dos distintar gráficas en uno solo
#adicionamos variable sex, tipo histograma, agregamos valores a histogramas ejes x y y
sns.jointplot(data=tips, x='total_bill', y='tip', hue='sex', kind='hist', marginal_ticks=True, marginal_kws=dict(bins=25,fill=False,multiple='dodge'))
plt.show()
```

```python
#Graficamos pairplot muestra variables numéricas, y las cruza
sns.pairplot(data=tips)
plt.show()
```

```python
#Graficamos pairplot muestra variables numéricas, y las cruza
#Agregamos variable sex, paleta coolwarm
sns.pairplot(data=tips, hue='sex', palette='coolwarm')
plt.show()
```

```python
#Graficamos pairplot muestra variables numéricas, y las cruza
#Agregamos variable sex, paleta coolwarm, quitamos información duplicada corner=True
sns.pairplot(data=tips, hue='sex', palette='coolwarm', corner=True)
plt.show()
```

```python
#Graficamos pairplot muestra variables numéricas, y las cruza
#Agregamos variable sex, paleta coolwarm, quitamos información duplicada corner=True
sns.pairplot(data=tips, hue='sex', palette='coolwarm', corner=True, kind='scatter')
plt.show()
```
