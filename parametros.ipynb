```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Creamos variable tips con data precargado de seaborn
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
#creamos grÃ¡fico desde seaborn
sns.displot(data=tips, x='total_bill')
plt.show()
```

```python
#creamos grÃ¡fico desde seaborn relaciÃ³n entre dos variables total de la propina y la propina
sns.displot(data=tips, x='total_bill', y='tip')
plt.show()
```

```python
#creamos grÃ¡fico desde seaborn relaciÃ³n tres variables total de la propina, la propina, sex
sns.displot(data=tips, x='total_bill', y='tip', hue='sex')
plt.show()
```

```python
#creamos grÃ¡fico desde seaborn relaciÃ³n  total de la propina,
sns.displot(data=tips, x='total_bill', hue='sex')
plt.show()
```

```python
#creamos grÃ¡fico tipo kde por densidad desde seaborn con total de la propina, sex
sns.displot(data=tips, x='total_bill', hue='sex', kind='kde')
plt.show()
```

```python
#creamos grÃ¡fico tipo kde por densidad desde seaborn con total de la propina, sex
#Sin leyenda
sns.displot(data=tips, x='total_bill', hue='sex', kind='kde', legend=False)
plt.show()
```

```python
#creamos grÃ¡fico tipo kde por densidad desde seaborn con total de la propina, sex
#Cambiamos paleta de colores a dark
sns.displot(data=tips, x='total_bill', hue='sex', kind='kde', legend='True', palette='dark')
plt.show()
```

```python
#creamos grÃ¡fico tipo kde por densidad desde seaborn con total de la propina, sex
#Cambiamos paleta de colores a dark
#Hacemos mÃ¡s transparente alpha=0.25
sns.displot(data=tips, x='total_bill', hue='sex', kind='kde', legend='True', palette='dark', alpha=0.25)
plt.show()
```
