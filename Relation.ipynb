```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Cargamos data frame seaborn llamado tips
tips = sns.load_dataset("tips")
tips.head(2) #Visualizamos solo las dos primeras filas del dataframe
```

```python
#Creamos gráfico scatterplot, entre totalbill y tip
sns.scatterplot(data=tips, x='total_bill', y='tip')
plt.show()
```

```python
#Creamos gráfico scatterplot, entre totalbill y tip, segmentamos por día
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day')
plt.show()
```

```python
#Creamos gráfico scatterplot, entre totalbill y tip, segmentamos por día,
#colocamos estilo por día x + o cuadrito
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time')
plt.show()
```

```python
#Creamos gráfico scatterplot, entre totalbill y tip, segmentamos por día,
#colocamos estilo por time x + o cuadrito
#Agregamos variable size, tamaño de comensales
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size')
plt.show()
```

```python
#Creamos gráfico scatterplot, entre totalbill y tip, segmentamos por día,
#colocamos estilo por día x + o cuadrito
#Agregamos variable size, tamaño de comensales
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size')
#movemos la leyenda
plt.legend(loc='center', bbox_to_anchor=(1.12,0.5))
plt.show()
```

```python
#Creamos gráfico scatterplot, entre totalbill y tip, segmentamos por día,
#colocamos estilo por día x + o cuadrito
#Agregamos variable size, tamaño de comensales
#movemos la leyenda
#modificamos los markers D diamonts s square para lunch almuerzo o dinner cena
markers = {"Lunch":"D", "Dinner":"s" }
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size', markers=markers)
plt.legend(loc='center', bbox_to_anchor=(1.12,0.5))
plt.show()
```

```python
#Creamos gráfico scatterplot, entre totalbill y tip, segmentamos por día,
#colocamos estilo por día x + o cuadrito
#Agregamos variable size, tamaño de comensales
#movemos la leyenda
#modificamos los markers D diamonts s square para lunch almuerzo o dinner cena
#cambiamos tamaño de gráfico
plt.figure(figsize=(8,8))
markers = {"Lunch":"D", "Dinner":"s" }
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size', markers=markers)
plt.legend(loc='center', bbox_to_anchor=(1.12,0.5))
plt.show()
```

```python
#Graficamos lineplot
sns.lineplot(data=tips, x='total_bill', y='tip')
plt.show()
```

```python
#Creamos gráfico lineplot, entre totalbill y tip, segmentamos por día,
#Agregamos variable size, tamaño de comensales
#movemos la leyenda
sns.lineplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size')
plt.legend(loc='center', bbox_to_anchor=(1.12,0.5))
plt.show()
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size')
plt.show()
```

```python
#Separamso en dos gráficos por columna='time'
sns.relplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size', col='time')
plt.show()
```
