```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Cargamos data frame de seaborn tips

tips = sns.load_dataset("tips")

tips.head(2) #mostramos solo las dos primeras filas del dataframe
```

```python
#Creamos gráfico seaborn con variable categórica sex

sns.countplot(data=tips, x='day', hue='sex')

plt.show()
```

```python
#Creamos gráfico seaborn con variable categórica sex, eje y

sns.countplot(data=tips, y='day', hue='sex')

plt.show()
```

```python
#Creamos gráfico dispersión seaborn con variable categórica sex, x day contra y totalbill

sns.stripplot(data=tips, x='day', y='total_bill', hue='sex')

plt.show()
```

```python
#Creamos gráfico dispersión seaborn con variable categórica sex, x day contra y totalbill
#propiedad dodge, unoa al lado del otro
sns.stripplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True)

plt.show()
```

```python
#Creamos gráfico dispersión seaborn con variable categórica sex, x day contra y totalbill
#propiedad dodge, unoa al lado del otro
#Fijamos antes un tamaño para el gráfico
plt.figure(figsize=(6,6))
sns.stripplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True)

plt.show()
```

```python
#Creamos gráfico "eje de dispersión" swarmplot, seaborn con variable categórica sex, x day contra y totalbill
#propiedad dodge, unoa al lado del otro
#Fijamos antes un tamaño para el gráfico
plt.figure(figsize=(6,6))
sns.swarmplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True)

plt.show()
```

```python
#Creamos gráfico boxplot seaborn con variable categórica sex, x day contra y totalbill
#propiedad dodge, unoa al lado del otro
#Fijamos antes un tamaño para el gráfico
plt.figure(figsize=(6,6))
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True)

plt.show()
```

```python
#Creamos gráfico boxplot con variable numérica
#Fijamos antes un tamaño para el gráfico
plt.figure(figsize=(6,6))
sns.boxplot(data=tips, y='total_bill')

plt.show()
```

```python
#Creamos gráfico boxplot y swarmplot seaborn con variable categórica sex, x day contra y totalbill
#propiedad dodge, unoa al lado del otro
#Fijamos antes un tamaño para el gráfico
plt.figure(figsize=(6,6))
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True)
sns.swarmplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True)
plt.show()
```

```python
#Creamos gráfico boxplot y swarmplot seaborn con variable categórica sex, x day contra y totalbill
#propiedad dodge, unoa al lado del otro
#Fijamos antes un tamaño para el gráfico
plt.figure(figsize=(6,6))
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True)
sns.swarmplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True, palette='dark:0')
plt.show()
```

```python
#Creamos gráfico boxplot y swarmplot seaborn con variable categórica sex, x day contra y totalbill
#propiedad dodge, unoa al lado del otro
#Fijamos antes un tamaño para el gráfico
plt.figure(figsize=(6,6))
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True)
sns.swarmplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True, palette='dark:0', marker='<')
plt.show()
```

```python
#Creamos gráfico violinplot seaborn con variable categórica sex, x day contra y totalbill
#propiedad dodge, unoa al lado del otro
#Fijamos antes un tamaño para el gráfico
plt.figure(figsize=(6,6))
sns.violinplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True)
plt.show()
```

```python
#Creamos gráfico violinplot seaborn con variable categórica sex, x day contra y totalbill
#propiedad dodge, unoa al lado del otro
#Fijamos antes un tamaño para el gráfico
plt.figure(figsize=(6,6))
sns.violinplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True, palette='pastel')
plt.show()
```

```python
#Creamos gráfico violinplot seaborn con variable categórica sex, x day contra y totalbill
#propiedad dodge, unoa al lado del otro
#Fijamos antes un tamaño para el gráfico
plt.figure(figsize=(6,6))
sns.violinplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True, palette='pastel', split=True)
plt.show()
```

```python
#Creamos gráfico violinplot seaborn con variable numérica

sns.violinplot(data=tips, y='total_bill')
plt.show()
```

```python
#Creamos gráfico catplot seaborn con variable categórica sex, x day contra y totalbill
sns.catplot(data=tips, x='day', y='total_bill', hue='sex')
plt.show()
```

```python
#Creamos gráfico catplot seaborn con variable categórica sex, x day contra y totalbill
#con propiedad dodge
sns.catplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True)
plt.show()
```

```python
#Creamos gráfico catplot seaborn con variable categórica sex, x day contra y totalbill
#con propiedad dodge, tip swarm
sns.catplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True, kind='swarm')
plt.show()
```

```python
#Creamos gráfico catplot seaborn con variable categórica sex, x day contra y totalbill
#con propiedad dodge, tip box
sns.catplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True, kind='box')
plt.show()
```

```python
#Creamos gráfico catplot seaborn con variable categórica sex, x day contra y totalbill
#con propiedad dodge, tip box
#Agregamos otra variable 'time' para graficar en otra columna, col='time'
sns.catplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True, kind='box', col='time')
plt.show()
```
