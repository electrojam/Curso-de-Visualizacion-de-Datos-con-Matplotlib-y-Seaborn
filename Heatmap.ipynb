```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")
tips.head(2)
```

```python
#llevamos tips a una estructura matricial especial
#obtenenmos la correlaci贸n entre las variables
tips.corr(numeric_only=True)
```

```python
#obtenenmos la correlaci贸n entre las variables pero en mapa de calor heatmap
sns.heatmap(tips.corr(numeric_only=True))
plt.show()
```

```python
#obtenenmos la correlaci贸n entre las variables pero en mapa de calor heatmap
#valores en mapa cada correlaci贸n
sns.heatmap(tips.corr(numeric_only=True), annot=True)
plt.show()
```

```python
#obtenenmos la correlaci贸n entre las variables pero en mapa de calor heatmap
#valores en mapa cada correlaci贸n, cambiamos colores
sns.heatmap(tips.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.show()
```

```python
#obtenenmos la correlaci贸n entre las variables pero en mapa de calor heatmap
#valores en mapa cada correlaci贸n, cambiamos colores, separamos l铆neas
sns.heatmap(tips.corr(numeric_only=True), annot=True, cmap='coolwarm', linewidth=5)
plt.show()
```

```python
#obtenenmos la correlaci贸n entre las variables pero en mapa de calor heatmap
#valores en mapa cada correlaci贸n, cambiamos colores, separamos l铆neas, l铆neas negras
sns.heatmap(tips.corr(numeric_only=True), annot=True, cmap='coolwarm', linewidth=5, linecolor='black')
plt.show()
```

```python
#obtenenmos la correlaci贸n entre las variables pero en mapa de calor heatmap
#valores en mapa cada correlaci贸n, cambiamos colores, separamos l铆neas, l铆neas negras
#modificamos escalara de valores
sns.heatmap(tips.corr(numeric_only=True), annot=True, cmap='coolwarm', linewidth=5, linecolor='black', vmin=0, vmax=1)
plt.show()
```

```python
#obtenenmos la correlaci贸n entre las variables pero en mapa de calor heatmap
#valores en mapa cada correlaci贸n, cambiamos colores, separamos l铆neas, l铆neas negras
#modificamos escalara de valores, quitamos barra
sns.heatmap(tips.corr(numeric_only=True), annot=True, cmap='coolwarm', linewidth=5, linecolor='black', vmin=0.5, vmax=1, cbar=False)
plt.show()
```
