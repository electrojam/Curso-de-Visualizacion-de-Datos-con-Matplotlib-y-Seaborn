```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
country = ['INDIA', 'JAPAN', 'MÉXICO', 'COLOMBIA', 'GERMANY'] #Creamos arreglo con 5 elementos
population = [1000, 800, 900, 1000, 300] #creamos otro arreglo con 5 datos
```

##RETO

```python
#creamos gráfica tipo barra entre country y population
plt.bar(country, population, width=0.5, color=['red', 'green'])
#obtenemos índice de arreglo y los cambiamos
plt.xticks(np.arange(5),('India','Japón', 'México', 'Colombia', 'Alemania'), rotation=45)
plt.title('Población por País')
plt.xlabel('País')
plt.ylabel('Población')
plt.show() #mostramos la gráfica
```

```python
#Graficamos de manera horizontal
plt.barh(country, population)
plt.show()
```
