```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = np.sin(x)
```

```python
#definimos objeto tipo figura con fig reemplaza fig=plt.figure(), y objeto tipo axes
fig, axes = plt.subplots()
axes.plot(x, y, 'b')
```

```python
#definimos objeto tipo figura con fig reemplaza fig=plt.figure(), y objeto tipo axes,
#en un plano de una fila y dos columnas
#donde axes devuelve un objeto tipo array al se puede ingresar
fig, axes = plt.subplots(nrows=1, ncols=2)
#graficamos en posión [0] del array axes
axes[0].plot(x,y, 'b')
#graficamos en posión [1] del array axes
axes[1].plot(y,x, 'r')
```

```python
#definimos objeto tipo figura con fig reemplaza fig=plt.figure(), y objeto tipo tupla ax1 y ax2,
#en un plano de una fila y dos columnas
#donde axes devuelve un objeto tipo array al se puede ingresar
fig, (ax1,ax2) = plt.subplots(nrows=1, ncols=2)
#graficamos
ax1.plot(x,y, 'b')
ax2.plot(y,x, 'r')
```

```python
#definimos objeto tipo figura con fig reemplaza fig=plt.figure(), y objeto tipo gráfico en un array,
#en un plano de una fila y dos columnas
#donde axes devuelve un objeto tipo array al se puede ingresar
fig, axes = plt.subplots(2, 4)
#graficamos en posición 0,0, eje x y eje y=cos(x)
axes[0,0].plot(x, np.cos(x))
#Graficamos los demás gráficos en distintas posiciones
axes[0,1].plot(x, np.sin(x), 'y')
axes[0,2].plot(x, np.tan(x), 'r')
axes[0,3].plot(x, np.cos(x)**2)
```

```python
#definimos objeto tipo figura con fig reemplaza fig=plt.figure(), y objeto tipo tupla ax1 hasta ax8,
#en un plano de una fila y dos columnas
#donde axes devuelve un objeto tipo array al se puede ingresar
fig, ((ax1, ax2, ax3, ax4), (ax5, ax6, ax7, ax8)) = plt.subplots(2, 4)
#graficamos en posición 0,0, eje x y eje y=cos(x)
ax1.plot(x, np.cos(x))
#Graficamos los demás gráficos en distintas posiciones
ax2.plot(x, np.sin(x), 'y')
ax3.plot(x, np.tan(x), 'r')
ax4.plot(x, np.cos(x)**2)
#aplicamos a objeto figura propiedad tidht_layout
fig.tight_layout()
```

```python
#definimos objeto tipo figura con fig reemplaza fig=plt.figure(), y objeto tipo tupla ax1 hasta ax8,
#en un plano de una fila y dos columnas
#donde axes devuelve un objeto tipo array al se puede ingresar
fig, ((ax1, ax2, ax3, ax4), (ax5, ax6, ax7, ax8)) = plt.subplots(2, 4)
#graficamos en posición 0,0, eje x y eje y=cos(x)
ax1.plot(x, np.cos(x))
#Graficamos los demás gráficos en distintas posiciones
ax2.plot(x, np.sin(x), 'y')
ax3.plot(x, np.tan(x), 'r')
ax4.plot(x, np.arcsin(x))
ax5.plot(x, np.arccos(x))
ax6.plot(x, np.arctan(x))
ax7.plot(x, np.sinh(x))
ax8.plot(x, np.cosh(x))
#aplicamos a objeto figura propiedad tidht_layout
fig.tight_layout()
```
