```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Creamos array de 11 elementos, valore entre 0 y 5
x = np.linspace(0,5,11)
#Creamos array con el seno de arreglo de x
y = np.sin(x)
```

```python
#Creamos objeto tipo figura fig y arreglo con dos gráficas axes, de 1x2
fig, axes = plt.subplots(1,2, figsize=(5,5)) #figsize=(5,5) tamaño de figura 5x5
axes[0].plot(x,y, label="$sin(x)$") #Graficamos en posición 1, definimos label
axes[0].set_title('Relacion X - Y') #Agregamos un título 'Relación X - Y'
axes[0].set_xlabel('X') #Agregamos etiqueta en eje X
axes[0].set_ylabel('Y') #Agregamos etiqueta en eje Y
axes[0].legend() #Mostramos el label="sin(x)"
axes[1].plot(y,x) #Graficamos en posición 2
axes[1].set_title('Relacion Y - X') #Agregamos un título 'Relación X - Y'
axes[1].set_xlabel('Y') #Agregamos etiqueta en eje Y
axes[1].set_ylabel('X') #Agregamos etiqueta en eje X
```

```python
#Hacemos el gráfico anterior pero sin orintado a objetos
plt.plot(x,y, label='sin(x)')
plt.title('Este es un título')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(loc='lower right', bbox_to_anchor=(1,0.2)) #Mostramos la leyenda y ubicamos abajo derecha
```
