```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.randint(1,50,100) #arreglo de 100 datos enteros entre 1-50
```

```python
plt.hist(data) #Dibujamos gráfica tipo histórico
plt.show() #mostramos gráfico
```

```python
plt.hist(data, bins=50, width=1) #Dibujamos gráfica tipo histórico,
#bins=número de barras distribución de datos, width=ancho de cada barra
plt.show() #mostramos gráfico
```

```python
plt.hist(data, bins=10, histtype='step') #Dibujamos gráfica tipo histórico,
#histtype='step' gáfico histótico tipo step
#bins=número de barras distribución de datos, width=ancho de cada barra
plt.show() #mostramos gráfico
```

##BOXPLOT

```python
plt.boxplot(data) #Creamos gráfico que muestra rango interpartídico
plt.show()
```

```python
#Creamos gráfico que muestra rango interpartídico, vertical, rellenado, con notch
 #el notch (enfoca datos hacia la mediana)
plt.boxplot(data, vert=False, patch_artist=True, notch=True)
plt.show()
```

```python
#Creamos gráfico que muestra rango interpartídico, vertical, rellenado, con notch
 #el notch (enfoca datos hacia la mediana)
 #Graficaremos con datos anómalos showfliers=True esto es por defecto
 #Si colocamos showfliers=False no muestra los datos anómalos y gráfica mostraría normal
 #con esto deja el rango 200 fuera del gráfico
data = np.append(data, 200)
plt.boxplot(data, vert=False, patch_artist=True, notch=True, showfliers=True)
plt.show()
```

##Scatter

```python
N = 50
#Creamos 5 arreglos tamaño de 50 elementos de datos aleatorios
x = np.random.rand(N)
y = np.random.rand(N)
area = (30 * np.random.rand(N)) ** 2
colors = np.random.rand(N)
```

```python
#Graficamos tipo scatter relación entre x - y
plt.scatter(x, y)
plt.show()
```

```python
#Graficamos tipo scatter, agregando tercera variable área (tamaño burbujas)
plt.scatter(x, y, s=area)
plt.show()
```

```python
#Graficamos tipo scatter, con tercera variable área (tamaño burbujas)
#Con cuarta variable colors
plt.scatter(x, y, s=area, c=colors)
plt.show()
```

```python
#Graficamos tipo scatter, con tercera variable área (tamaño burbujas)
#Con cuarta variable colors, con quinta variable marker marcador
plt.scatter(x, y, s=area, c=colors, marker='x')
plt.show()
```

```python
#Graficamos tipo scatter, con tercera variable área (tamaño burbujas)
#Con cuarta variable colors, con variable marker marcador
#Con variable alpha transparencia de los markers
plt.scatter(x, y, s=area, c=colors, marker='o', alpha=0.5)
plt.show()
```
